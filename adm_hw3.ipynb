```python
linkfiles = open("/content/drive/MyDrive/adm_hw3/place_lists.txt", "w")

for i in tqdm(range(1,401)): 
  url = 'https://www.atlasobscura.com/places?page=' + str(i) + '&sort=likes_count'
  response = requests.get(url)
  soup = BeautifulSoup(response.content, features = 'lxml')
  for a in soup.find_all('a', class_ ='col-md-3 nav-card'):
    linkfiles.write(a.get('href')+ '\n')

linkfiles.close()
```

```python
for page in tqdm(range(1,401)):
  subfolder = "/content/drive/MyDrive/adm_hw3/Html_page/page_{}".format(page)
  os.makedirs(subfolder)

  f= open("/content/drive/MyDrive/adm_hw3/place_lists.txt", 'r', encoding = "utf8")
  lines = f.readlines()[(page-1)*18:(page)*18]
  lines = [line.rstrip() for line in lines]
  f.close

  i=1+18*(page-1)
  #per ogni link che abbiamo letto creamo un nuovo .html file 
  for link in lines: 
    link = 'https://www.atlasobscura.com'+ link
    html = requests.get(link)

    file_name = '{}/{}.html'.format(subfolder, i)
    g= open(file_name, 'w', encoding= "utf8")
    g.write(html.text)
    i+=1
```
